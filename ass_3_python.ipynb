## Ass. 3 - Kiva Loan Analysis

This notebook analyzes the Kiva loan dataset using Pandas.
The goal is to create 6 meaningful questions and answer them using:
- row selection using string/numerical conditions,
- sorting with `.sort_values()`,
- grouping and aggregation,
- and creating a new variable using `apply` and `lambda`.

## Q1 — Which countries received the highest total loan amount?

Here we group the data by country, sum all loan amounts, and display the top 10.
This uses:

- `groupby()`
- `sum()`
- `sort_values()`

This answers: *Which countries get the most funding through Kiva?*

```python
total_by_country = (
    df.groupby("country")["loan_amount"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

total_by_country
```

## Q2 — What is the average loan amount per economic sector?

We group loans by sector and compute the mean loan size.
This lets us understand which sectors receive larger or smaller loans on average.

Uses:
- `groupby()`
- `mean()`
- `sort_values()`

```python
avg_by_sector = (
    df.groupby("sector")["loan_amount"]
      .mean()
      .sort_values(ascending=False)
)

avg_by_sector
```

## Q3 — Which loans are above the global average loan amount?

We first calculate the overall average (`mean()`) and then filter all rows where
the loan amount is greater than that. This uses a *numerical condition*.

This shows which loans are above-average compared to the dataset overall.


```python
mean_loan = df["loan_amount"].mean()
above_avg_loans = df[df["loan_amount"] > mean_loan]

above_avg_loans.head()
```

## Q4 — Show all loans from the Agriculture or Food sectors.

This uses a **string-based filter** with `.isin()`.  
It allows selecting specific categories from a text column.

This answers: *How many loans are related to food or farming?*


```python
filtered_sectors = df[df["sector"].isin(["Agriculture", "Food"])]
filtered_sectors.head()
```

## Q5 — Sort loans by term (ascending) and loan amount (descending).

This uses `sort_values()` with two columns.  
It answers: *Which loans are short-term but have large amounts?*

```python
sorted_loans = df.sort_values(
    by=["term_in_months", "loan_amount"],
    ascending=[True, False]
)

sorted_loans.head()
```

## Q6 — Create a new variable categorizing loans as Small, Medium, or Large.

This uses `apply()` and a `lambda` function.  
It demonstrates how to create new columns based on conditions.

Rules:
- Small  < 500
- Medium < 1000
- Large  >= 1000

```python
df["loan_size_category"] = df["loan_amount"].apply(
    lambda x: "Small" if x < 500 else ("Medium" if x < 1000 else "Large")
)

df[["loan_amount", "loan_size_category"]].head()
```
